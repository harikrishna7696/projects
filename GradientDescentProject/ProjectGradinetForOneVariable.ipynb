```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data = np.loadtxt('../data.csv', delimiter=',')
data.shape
X = data[:,0].reshape(-1,1) 
Y = data[:,1]
X.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.20)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
alg = LinearRegression()
alg
```

```python
alg.fit(X_train, Y_train)
Y_pred = alg.predict(X_test)
Y_pred
```

```python
d = {'Y_test':Y_test, 'Y_pred':Y_pred}
```

```python
df_pred = pd.DataFrame(d,index=None)
```

```python
df_pred
```

```python
def step_gradient(data, learning_rate, m, c):
    m_slope = 0
    c_slope = 0
    N = len(data)
    for i in range(N):
        X = data[i, 0]
        Y = data[i, 1]
        m_slope += (-2/N) * (Y - m * X - c) *(X)
        c_slope += (-2/N) * (Y - m * X - c)
        new_m = m - learning_rate * m_slope
        new_c = c - learning_rate * c_slope
    return new_m, new_c
```

```python
def cost(data, m, c):
    total_cost = 0
    M = len(data)
    for i in range(M):
        x = data[i, 0]
        y = data[i, 1]
        total_cost += (1/M) * ((y - m * x - c)**2)
    return total_cost
```

```python
def gradient(data, learning_rate, num_of_iteration):
    m = 0
    c = 0
    for i in range(num_of_iteration):
        m, c = step_gradient(data, learning_rate, m, c)
        print(i, 'cost: ', cost(data,m, c))
    return m, c
```

```python
def predict(X_train, m, c):
    Y_gradint_pred = []
    N = len(X_train)
    for i in range(N):
        X = X_train[i]
        Y_gradint_pred.append(*(m*X)+c)
    return Y_gradint_pred
```

```python
def run(n, iterations):
    data = np.loadtxt('../data.csv', delimiter=',')
    learning_rate = n
    num_of_iteration = iterations
    m, c = gradient(data,learning_rate,num_of_iteration)
    return m,c, learning_rate
```
