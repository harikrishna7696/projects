### 1) Login to your Instagram Handle

```python
def LoginInstagram(username, password):
    try:
        # It waits until condition becomes true in given 8 sec if page takes more than time to load it's thrown error
        wait.until(EC.presence_of_element_located((by.NAME, 'username'))).send_keys(username)
        # It waits until condition becomes true in given 8 sec if page takes more than time to load it's thrown error
        wait.until(EC.presence_of_element_located((by.NAME, 'password'))).send_keys(password)
        wait.until(EC.element_to_be_clickable((by.TAG_NAME, 'button'))).submit()
        
        # logging is taking time so that's why here give sleep time
        time.sleep(5)
        driver.find_element(by.CLASS_NAME, '_acan._acao._acas').click()
        print('Logged In SuccessFully')
    except Exception:
        print('error')
        
```

### 4. Follow/Unfollow given handle - 

```python
def follow(username):
    try:
        # this function take username and open the profile
        search_and_open(username)
        time.sleep(2)
        # finding the follow button
        follow_btn = driver.find_element(by.CLASS_NAME, '_aacl._aaco._aacw._aad6._aade')
        if follow_btn.text == 'Follow':
            # clicking on the follow button if we not followed
            follow_btn.click()
            print("Now You are following: ", username)
        else:
            # if we already followed just return this message 
            print("You have already following: ", username)
    except Exception:
        print("Something Went Wrong")
```

```python
def unfollow(username):
    try:
        search_and_open(username)
        time.sleep(2)
        follow_btn = driver.find_element(by.CLASS_NAME, '_aacl._aaco._aacw._aad6._aade')
        if follow_btn.text == 'Following':
            follow_btn.click()
            time.sleep(3)
            driver.find_element(by.CLASS_NAME, '_a9--._a9-_').click()
            print("Now You are Unfollowed: ", username)
        else:
            print("You Already Unfollowed: ", username)
    except Exception:
        print('Something went wrong.')
```

### 5. Like/Unlike posts

```python
def like(username):
    try:
        # search and open profile 
        search_and_open(username)
        # clicking on the first post from profile
        wait.until(EC.presence_of_element_located((by.CLASS_NAME, '_aabd._aa8k._aanf'))).click()
        for i in range(1, 31):
            # finding the like button
            Like = wait.until(EC.presence_of_element_located((by.XPATH,'//span[@class="_aamw"]/button')))
            GetText = bs(Like.get_attribute('innerHTML'),'html.parser').svg['aria-label']
            
            # if text is 'Like'
            if GetText == 'Like':
                print(f'Liked post->{i}')
                # clicking on the like button
                Like.click()
            else:
                # if already clicked return this message
                print(f"You have already Liked. post->{i}")
            time.sleep(5)
            driver.find_element(by.CLASS_NAME, '_aaqg._aaqh').click()
    except KeyboardInterrupt:
        # if you stoppen then this message will return
        print("You Stopped")
    except Exception:
        print("Something Went Wrong...!")
```

```python
# 5.1 Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
like('dilsefoodie')
```

### Extract list of followers

```python
def followers_list():
    try:
        driver.find_element(by.XPATH, '//*[@id="mount_0_0_fm"]/div/div/div/div[1]/div/div/div/div[1]/div[2]/div[2]/section/main/div/header/section/ul/li[2]').click()
        frame = wait.until(EC.presence_of_element_located((by.XPATH,'//div[@class="_aano"]')))
        for i in range(50):
                time.sleep(2)
                driver.execute_script('arguments[0].scrollTop=arguments[0].scrollHeight',frame)
        li = driver.find_elements(by.CLASS_NAME, '_ab8y._ab94._ab97._ab9f._ab9k._ab9p._abcm')
        followers = list()
        for i in li:
            followers.append(i.text)
        return followers
    except Exception:
        print("Something went wrong.")
```

#### 7. Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
### 1. If You have already seen the story.
### 2. Or The user has no story.
### 3. Or View the story if not yet seen. 

```python
def check_story():
    try:
        # Locating the story
        story = wait.until(EC.presence_of_element_located((by.CLASS_NAME,'_aarf._aarg')))
        
        #check the Profile photo size to find out story is available or not
        height = wait.until(EC.presence_of_element_located((by.CLASS_NAME,'_aarh'))).get_attribute('height')
        # checking height of the story
        if height =='208':
            # if the story has already seen
            print('Already seen the story')
        else:
            # viewing the story
            print("Viewing the story")
            story.click()
    except Exception:
        
        # Getting error it means No story is available
        print('No story is available to view')
```
